```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import dataset

```python
dataset = pd.read_csv('machinelearning_dataset.csv')
```

```python
dataset["Cuisines"].fillna("unknown")
rows,columns=dataset.shape
dataset.head()
```

```python
boolean_list = ["Has Table booking", "Has Online delivery", "Is delivering now", "Switch to order menu"]

# Define a function to replace values
def boolean_replace(value):
    if value == "Yes":
        return 1
    elif value == "No":
        return 0
    else:
        return value  # Return the original value if it's not "Yes" or "No"

# Apply the function to selected columns
dataset[boolean_list] = dataset[boolean_list].applymap(boolean_replace)

# Display the modified DataFrame
dataset.head()
```

```python
#here some columns which has some invalid characters "?"
```

```python
accented_columns = ["Restaurant Name", "city", "Address","Locality","Locality Verbose","Cuisines"]
```

```python
accented_columns = ["Restaurant Name", "City", "Address","Locality","Locality Verbose","Cuisines"]
from unidecode import unidecode
def clean_text(text):
    if isinstance(text, str):
        return unidecode(text)
    return text
dataset[accented_columns] = dataset[accented_columns].applymap(clean_text)
dataset.tail()
```

```python
dataset.columns
dataset.head()
```

```python
dataset.drop("Rating color",axis = 1,inplace=True)
```

```python
rating_scale = {
    'Excellent':10,
    'Very Good':8,
    'Good'     :7,
    'Average'  :5,
    'Not rated':0,
    'Poor'     :2
}
for i in range(rows):
    if dataset.loc[i,"Rating text"] in rating_scale.keys():
        dataset.loc[i,"Rating text"]=rating_scale[dataset.loc[i,"Rating text"]]
dataset["Rating text"].astype(int)
dataset.head()
dataset.to_csv("cleaned1.csv", index=False)
```
